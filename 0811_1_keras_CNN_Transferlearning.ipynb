```
import keras
import numpy as np
from keras.applications import vgg16, inception_v3, mobilenet
```

```
from keras.preprocessing.image import load_img, img_to_array
import matplotlib.pyplot as plt
```

```
vgg_model = vgg16.VGG16(weights='imagenet') # imagenet : image classification
```

```
from google.colab import files
uploaded = files.upload()
```

```
load_img('pic.jpg')
```

```
filename = 'pic.jpg'
org = load_img(filename, target_size=(224,224))
img = img_to_array(org)
plt.imshow(np.uint8(img))
```

```
img.shape
```

```
img[0][0]
```

```
x = np.expand_dims(img, axis=0)
x.shape
```

```
x = vgg16.preprocess_input(x)
pred = vgg_model.predict(x)
print(pred)
```

```
from keras.applications.imagenet_utils import decode_predictions
label = decode_predictions(pred)
print(label)
```
